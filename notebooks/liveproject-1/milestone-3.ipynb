```python
import json
from collections import Counter
from src.preprocessing.prepro import TextPrep
```

```python
with open('data/data_tfidf.json', 'r') as j:
    data = json.load(j)

data[0]
```

```python
full_text = []
_ = [full_text.extend(wikipedia["tokenized_text"]) for wikipedia in data]
full_text_counter = Counter(full_text)
full_text_tokens = [token for token in full_text_counter]
full_text_tokens[:10]
```

```python
with open('data/features.json', 'r') as j:
    features_order = json.load(j)

features_order
```

```python
# creating an inverted index
result = {}
tfidf_value = []
for token in full_text_tokens:
    doc = []
    for wikipedia in data:
        if token in wikipedia["tokenized_text"]:
            print(token)
            doc.append(wikipedia["title"])
            result[token] = doc
            tfidf_value.append([
                wikipedia["tfidf_vec"][i]
                for i, token_ in enumerate(wikipedia["tokenized_text"])
                if token == token_
            ])
    break

tfidf_value
```

```python
query = "zoonotic diseases"

# preprocessing
text_prep = TextPrep(query)
text_prep.lowercasing()
text_prep.get_lemmas()
query_tokens = text_prep.final_tokens
```

```python
docs = []
for token_idx, doc in result.items():
    for query_token in query_tokens:
        if query_token == token_idx:
            docs.extend(doc)

set(docs)
```
