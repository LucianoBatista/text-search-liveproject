## Word Vectors

The goal of word vectors is to achieve numerical understanding of language so that a computer can perform more complex tasks on that corpus. Like understanding synonyms or similarities.

Synonym substitution, a common method of data augmentation, does not take into account syntactical differences of synonyms.

Word Vectors are normally trained on specific tasks and tend to understanding similar occurrences (synonym) of words.

```python
# word embedding of that token
token1.vector
```

```python
your_word = "country"

ms = nlp.vocab.vectors.most_similar(
    np.asarray([nlp.vocab.vectors[nlp.vocab.strings[your_word]]]), n=10
)
words = [nlp.vocab.strings[w] for w in ms[0][0]]
distances = ms[2]
print(words)
```

## spaCy's EntityRuler

```python
text = 'West Chestertenfieldville was referenced in Mr. Deeds.'
doc = nlp2(text)
```

```python
for ent in doc.ents:
    print("{:30} | {}".format(ent.text, ent.label_))
```

```python
ruler = nlp2.add_pipe("entity_ruler")
```

```python
nlp2.analyze_pipes()
```

```python
patterns = [
    {"label": "GPE", "pattern": "West Chestertenfieldville"}
]
```

```python
ruler.add_patterns(patterns)
```

```python
doc2 = nlp2(text)

for ent in doc2.ents:
    print("{:30} | {}".format(ent.text, ent.label_)) 
```

```python
nlp2 = spacy.load("en_core_web_lg")

patterns = [
    {"label": "GPE", "pattern": "West Chestertenfieldville"},
    {"label": "FILM", "pattern": "Mr. Deeds"}
]

ruler = nlp2.add_pipe('entity_ruler', before="ner")
ruler.add_patterns(patterns)
```

```python
doc2 = nlp2(text)

for ent in doc2.ents:
    print("{:30} | {}".format(ent.text, ent.label_)) 
```

## Matcher

```python
nlp = spacy.load('en_core_web_lg')
matcher = Matcher(nlp.vocab)
pattern = [
    {"LIKE_EMAIL": True}
]

matcher.add('EMAIL_ADDRESS', [pattern])
doc = nlp('This is an email address: luciano@gmail.com')
matches = matcher(doc)

print(matches)
```

```python
print(nlp.vocab[matches[0][0]].text)
```

```python
list(doc)[matches[0][1]:matches[0][2]]
```

```python
with open('data/wiki_mlk.txt', 'r') as f:
    data = f.read()

data
```

```python
nlp = spacy.load('en_core_web_lg')
matcher = Matcher(nlp.vocab)

pattern = [
    {"POS": "PROPN"}
]

matcher.add('PROPER_NOUN', [pattern])
doc = nlp(data)
matches = matcher(doc)

for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

```python
spacy.displacy.render(doc, style='ent')
```

```python
nlp = spacy.load('en_core_web_lg')
matcher = Matcher(nlp.vocab)

pattern = [
    {"POS": "PROPN", "OP": "*"}
]

matcher.add('PROPER_NOUN', [pattern])
doc = nlp(data)
matches = matcher(doc)

for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

```python
nlp = spacy.load('en_core_web_lg')
matcher = Matcher(nlp.vocab)

pattern = [
    {"POS": "PROPN", "OP": "*"}
]

matcher.add('PROPER_NOUN', [pattern], greedy='LONGEST')
doc = nlp(data)
matches = matcher(doc)

for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

```python
nlp = spacy.load('en_core_web_lg')
matcher = Matcher(nlp.vocab)

pattern = [
    {"POS": "PROPN", "OP": "+"},
    {"POS": "VERB"}
]

matcher.add('PROPER_NOUN', [pattern], greedy='LONGEST')
doc = nlp(data)
matches = matcher(doc)
matches.sort(key = lambda x: x[1])

for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

## RegEx in spaCy

```python
nlp = spacy.load("en_core_web_lg")
contextualSpellCheck.add_to_pipe(nlp)
doc = nlp('Income was $9.4 milion compared to the prior year of $2.7 milion.')
print(doc._.performed_spellCheck)
print(doc._.outcome_spellCheck)
```

```python
from spellchecker import SpellChecker

portuguese = SpellChecker(language='pt')

# find those words that may be misspelled
misspelled = portuguese.unknown(['escol', 'estudar', 'mar', 'oinbos'])

for word in misspelled:
    # Get the one `most likely` answer
    print(portuguese.correction(word))

    # Get a list of `likely` options
    print(portuguese.candidates(word))
```
